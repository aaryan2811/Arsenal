```python
from google.colab import drive
drive.mount('/content/drive')
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
filename = '/content/Week51.xlsx'
df = pd.read_excel(filename)
df.head()
```

```python
year = df['Season'].values
final = df['Final Position'].values
plt.style.use('seaborn')
plt.plot(year, final,marker = '.', markersize = 10, c = 'b')
plt.gca().invert_yaxis()
plt.xlabel('Year', fontsize=15)
plt.ylabel('Final Position in League Table', fontsize=15)
plt.title('Arsenal Final Position in League Table Over the Years', fontsize=20)
plt.xticks(rotation = 45)
```
